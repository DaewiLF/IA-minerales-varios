```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import os


!unzip -q minet.zip -d dataset/

DIRECTORIO_DATOS = '/content/dataset/minet'
TAMAÑO_IMAGEN = (224, 224)
TAMANO_LOTE = 32

print("Librerías cargadas y dataset descomprimido.")
```

```python

datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    brightness_range=[0.8, 1.2],
    horizontal_flip=True,
    validation_split=0.2
)

print("Cargando conjunto de Entrenamiento")
train_generator = datagen.flow_from_directory(
    DIRECTORIO_DATOS,
    target_size=TAMAÑO_IMAGEN,
    batch_size=TAMANO_LOTE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

print("\nCargando conjunto de Validación")
val_generator = datagen.flow_from_directory(
    DIRECTORIO_DATOS,
    target_size=TAMAÑO_IMAGEN,
    batch_size=TAMANO_LOTE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

clases_minerales = list(train_generator.class_indices.keys())
print(f"\nClases detectadas ({len(clases_minerales)}): {clases_minerales}")
```

```python
base_model = ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False


x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.4)(x)
predicciones = Dense(len(clases_minerales), activation='softmax')(x)

modelo = Model(inputs=base_model.input, outputs=predicciones)


modelo.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Arquitectura del modelo lista.")
```

```python
parada_temprana = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

EPOCAS = 25

print("Iniciando entrenamiento...")
historial = modelo.fit(
    train_generator,
    epochs=EPOCAS,
    validation_data=val_generator,
    callbacks=[parada_temprana]
)

print("\n¡Entrenamiento completado!")
```

```python
# Gráficos de evolución
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(historial.history['accuracy'], label='Entrenamiento', color='blue')
plt.plot(historial.history['val_accuracy'], label='Validación', color='orange')
plt.title('Evolución de la Exactitud (Accuracy)')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(historial.history['loss'], label='Entrenamiento', color='blue')
plt.plot(historial.history['val_loss'], label='Validación', color='orange')
plt.title('Evolución de la Pérdida (Loss)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 1. Generar predicciones para la Matriz de Confusión
Y_pred = modelo.predict(val_generator)
y_pred_classes = np.argmax(Y_pred, axis=1)
y_true = val_generator.classes

# 2. Dibujar la Matriz de Confusión
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=clases_minerales, yticklabels=clases_minerales)
plt.title('Matriz de Confusión de Minerales')
plt.ylabel('Etiqueta Real')
plt.xlabel('Predicción de la IA')
plt.show()

# 3. Reporte de Métricas (Precisión, Sensibilidad, F1-Score)
print("\nReporte de Clasificación:")
print(classification_report(y_true, y_pred_classes, target_names=clases_minerales))

# 4. Guardar el modelo en formato estándar y formato Lite (Edge Computing)
modelo.save('modelo_minerales.h5')

converter = tf.lite.TFLiteConverter.from_keras_model(modelo)
tflite_model = converter.convert()

with open('modelo_minerales.tflite', 'wb') as f:
  f.write(tflite_model)

print("\nModelos guardados: 'modelo_minerales.h5' y 'modelo_minerales.tflite' (Inferencia local).")
```
